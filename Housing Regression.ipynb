```python
from sklearn.preprocessing import MinMaxScaler as mms
```

```python
mmsa=mms()
```

```python
mmsa.fit(x)
```

```python
x=mmsa.transform(x)
```

```python
x=pd.DataFrame(x)
```

```python
x.columns=['House_size','Number_of_rooms','Year_of_construction']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train, x_test, y_train , y_test=tts(x,y,test_size=0.3,random_state=101)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model_dl=Sequential()
```

```python
model_dl.add(Dense(4,activation='relu'))
model_dl.add(Dense(1))
```

```python
model_dl.compile(optimizer='rmsprop',loss='mse')
model_dl.fit(x_train,y_train,epochs=500)
```

```python
pd.DataFrame(model_dl.history.history).plot()
```

```python
model_dl.evalute(x_test,y_test,verbose=0)
```
